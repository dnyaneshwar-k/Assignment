```python
#considering different genders mean different patients even for same full name,dob
def input_dataset1(data):
    m=len(data)
    for index,row in data.iterrows():
        dob_selected=row["dob"]
        g_selected=row["gn"]
        fn_selected=row["full_name"]
        if(row["duplicate"]!=False):
            for j in range(index+1,m):
                rc=data.iloc[j]
                dob_compare=rc["dob"]
                g_compare=rc["gn"]
                fn_compare=rc["full_name"]
                if(g_selected==g_compare):
                    if(fuzz.token_set_ratio(fn_selected,fn_compare)>90):
                        data.set_value(j,"duplicate",False)
    data=data[data["duplicate"]==True]
    data=data.reset_index(drop=True)
    print(data)
    print(data.shape)
    data=data.drop(["full_name","duplicate"],axis=1)
    data.to_csv("output_unique_gender.csv", index=False)
```
